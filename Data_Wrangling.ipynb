## Data Preparation

### Selecting the columns

```python
definitive_drop = ['ECONOMIC_GROUP_ID', 'FLAG_IMPARIDADE', 'MEDIA_GER', 'PIB_VAR', 'MEDIA_NOTA_SEGM']
data = data.drop(columns= definitive_drop)
```

Now, we create another dataset $data_p$ which won't contain any outcome variable from the other column

```python
# Select columns to drop from the model training process 
columns_to_drop = [
    'PD_EST',
    'PD_CALIBRADA_EST',
    'PD_AJUST_EST',
    'PD_AJUST_QL',
    'SCORE_QL',
    'SCORE_EST',
    'SCORE_AJUST_QL',
    'SCORE_AJUST_EST',
    'SCORE_FINAL',
    'PD_CALIBRADA_FINAL',
    'PD_FINAL',
    'RATING_ESTATISTICO',
    'RATING_ESTATISTICO_STANDARD',
    'FINAL_RATING_INICIAL',
    'RATING_FILTROS',
    'RATING_EST_FILTROS',
    'FINAL_RATING',
    'FINAL_RATING_EST',
    'FINAL_RATING_NO_DEFAULT',
    'FINAL_RATING_EST_STANDARD',
    'FINAL_RATING_NO_DEFAULT_STANDARD']


# Drop selected columns from the DataFrame
data_p = data.drop(columns=columns_to_drop)

# Convert the 'LOAD_DTTM' column to a date
data_p['LOAD_DTTM'] = pd.to_datetime(data_p['LOAD_DTTM'])
```

## Filtering


```python
# Filter rows where 'LOAD_DTTM' is not equal to '2019-07-24' because this was an system error
data_p = data_p[data_p['LOAD_DTTM'] != '2019-07-24']
```

### Dealing with the NaN's
The maximum proportion of NaN's per row

```python
threshold_proportion = 0.7  # For example, 30%

# Calculate the minimum number of non-NaN values required based on the proportion
threshold = int((1 - threshold_proportion) * len(data_p.columns))

# Exclude rows with a proportion of NaN values greater than the threshold
data_p = data_p.dropna(thresh=threshold)

num_rows_meeting_threshold = len(data_p.dropna(thresh=threshold))

print("Number of rows meeting the threshold:", num_rows_meeting_threshold)
```

### Removing the duplicates
We start by removing the duplicated rows

```python
# Count the number of duplicate rows
num_duplicates = data_p.duplicated(keep='first').sum()
print("Number of Duplicate Rows:", num_duplicates)
```

```python
data_p = data_p.drop_duplicates()
```

We encounter situations where the same 'IDENTIFIER' is associated twice with the same 'LOAD_DTTM,' and the corresponding rows are not identical due to variations in the 'SECTION_ECONOMIC_ACTIVITY_CD.' In such cases, we remove rows with 'NAN' values in the 'SECTION_ECONOMIC_ACTIVITY_CD'. 

```python
# Defining a custom filter function
def custom_filter(group):
    if len(group) == 1:
        return group
    else:
        if len(group['CLASSIFICATION_Q'].unique()) > 1:
            if len(group.dropna(subset=['CLASSIFICATION_Q'])) == 1:
                return group.dropna(subset=['CLASSIFICATION_Q'])
            else:       #len(group.dropna(subset=['CLASSIFICATION_Q'])) > 1
                min_value = group['CLASSIFICATION_Q'].min()
                min_rows = group[group['CLASSIFICATION_Q'] == min_value]
        elif len(group['ON_BALANCE_AMT'].unique()) > 1:
            return group.dropna(subset=['ON_BALANCE_AMT'])
        elif len(group['SECTION_ECONOMIC_ACTIVITY_CD'].unique()) > 1:
            return group.dropna(subset=['SECTION_ECONOMIC_ACTIVITY_CD'])


# Apply the custom filter to each group
data_p = data_p.groupby(['IDENTIFIER', 'LOAD_DTTM']).apply(custom_filter).reset_index(drop=True)
data_p = data_p.drop('ON_BALANCE_AMT', axis = 1)
# Check for duplicated rows
len(data_p[data_p.duplicated(subset=['IDENTIFIER', 'LOAD_DTTM'])])
# data_p = data_p.drop_duplicates()
```

### Filter for SME's with less than 12 months of observations


```python
# Calculate the number of months of observations per 'IDENTIFIER'
temp = data_p.groupby('IDENTIFIER')['LOAD_DTTM'].count().reset_index()
temp.columns = ['IDENTIFIER', 'n_months_obs']

# Left join 'n_months_obs' back to the DataFrame
data_p = data_p.merge(temp, on='IDENTIFIER', how='left')

# Calculate the maximum 'DEFAULT_FLG' per 'IDENTIFIER'
temp = data_p.groupby('IDENTIFIER')['DEFAULT_FLG'].max().reset_index()
temp.columns = ['IDENTIFIER', 'ever_default']

# Left join 'ever_default' back to the DataFrame
data_p = data_p.merge(temp, on='IDENTIFIER', how='left')

# Filter rows where 'n_months_obs' is greater than or equal to 12
data_p = data_p[data_p['n_months_obs'] >= 12]

# Drop the 'n_months_obs' column
data_p = data_p.drop(columns=['n_months_obs'])
```

### Creating the $Target$ variable

```python
import pandas as pd

data_p['LOAD_DTTM'] = pd.to_datetime(data_p['LOAD_DTTM'])
data_p = data_p.sort_values(by=['IDENTIFIER', 'LOAD_DTTM'])
# Initialize an empty list to store the TARGET values
target_values = []
# Iterate over rows
for index, row in data_p.iterrows():
    identifier = row['IDENTIFIER']
    load_dttm = row['LOAD_DTTM']
    default_flg = row['DEFAULT_FLG']
    # Filter rows within the next 12 months for the same identifier
    next_12_months = data_p[(data_p['IDENTIFIER'] == identifier) & (data_p['LOAD_DTTM'] > load_dttm) 
                            & (data_p['LOAD_DTTM'] <= load_dttm + pd.DateOffset(months=12))]
    # Check if there is any row with DEFAULT_FLG being 1 within the next 12 months
    target = int((next_12_months['DEFAULT_FLG'] == 1).any())
    # Append the result to the list
    target_values.append(target)
# Add the 'TARGET' column to the DataFrame
data_p['TARGET'] = target_values
# Reset the index
data_p = data_p.reset_index(drop=True)
```
